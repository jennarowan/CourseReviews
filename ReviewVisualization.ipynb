```python
import pandas
from datetime import datetime
from pytz import utc
import matplotlib.pyplot as plt
data = pandas.read_csv("reviews.csv", parse_dates=["Timestamp"])
```

```python
data.head()
```

```python
data["Day"] = data["Timestamp"].dt.date
day_average = data.groupby(["Day"]).mean()
day_average.head()
```

```python
day_average.index
```

```python
plt.figure(figsize = (25, 5))
plt.plot(day_average.index, day_average["Rating"])
```

#### Rating average by week

```python
data["Week"] = data["Timestamp"].dt.strftime("%Y-%U")
data
```

```python
week_average = data.groupby(["Week"]).mean()
plt.figure(figsize = (25, 5))
plt.plot(week_average.index, week_average["Rating"])
```

#### Ratings average by month

```python
data["Month"] = data["Timestamp"].dt.strftime("%Y-%m")
data
```

```python
month_average = data.groupby(["Month"]).mean()
plt.figure(figsize = (25, 5))
plt.plot(month_average.index, month_average["Rating"])
```

### Average rating by month by course

```python
month_average_by_course = data.groupby(["Month", "Course Name"])["Rating"].mean().unstack()
month_average_by_course.plot(figsize = (25, 8))
```

```python
month_average_by_course
```

#### Find day of week with highest ratings

```python
data["Day of Week"] = data["Timestamp"].dt.strftime("%w")
data["Day of Week"] = data["Day of Week"].replace(["0", "1", "2", "3", "4", "5", "6"],
     ["Sunday", "Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday"])
day_order = ["Sunday", "Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday"]
day_of_week_average = data.groupby(["Day of Week"]).mean()
day_of_week_average.loc[day_order].plot(figsize = (25, 8))
```

```python
# Better version
data["Day of Week"] = data["Timestamp"].dt.strftime("%A")
day_order = ["Sunday", "Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday"]
day_of_week_average = data.groupby(["Day of Week"]).mean()
day_of_week_average.loc[day_order].plot(figsize = (15, 3))
```
